```python
x = []
y = []
for addr in animal[''] :
    address = addr
    url = f"https://dapi.kakao.com/v2/local/search/address.json?query={address}"
    result = requests.get(urlparse(url).geturl(), headers={"Authorization":"KakaoAK 7880cc9dfabd083df48445c16d075a26"}).json()
    try :
        x.append(result['documents'][0]['x'])
        y.append(result['documents'][0]['y'])
    except :
        x.append('-')
        y.append('-')
```

```python
animal['x'] = x
```

```python
animal['y'] = y
```

```python
kkma = Kkma()
```

```python
special_list = []

for i in tqdm_notebook(range(len(animal))):
    if animal[''][i].replace(' ','') == '':
        special_list.append('')
    else :
        pos_character = kkma.pos(animal[''][i])

        word_list = []

        for pos in pos_character:
            if pos[1] in ['NR','NNM','XR','NNG','NNP', 'VV', 'VA', 'MAG']:
                word_list.append(pos[0])
                
        word = " ".join(word_list)

        special_list.append(word)
```

```python
animal['SPECIAL_WORD'] = special_list
```
