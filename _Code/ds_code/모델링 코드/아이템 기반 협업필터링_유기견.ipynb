```python
# 수식을 이용해서 개인화된 예측 평점을 구하는 함수
# 개인화된 예측 평점을 계산하고 정규화
def predict_rating(ratings_arr, item_sim_arr):
    scores_pred = ratings_arr.dot(item_sim_arr) / np.array([np.abs(item_sim_arr).sum(axis=1)])
    return scores_pred
```

```python
scores_pred = predict_rating(scores_matrix.values,item_sim_df.values)
scores_pred_matrix = pd.DataFrame(data=scores_pred, index= scores_matrix.index,
                                   columns = scores_matrix.columns)
print(scores_pred_matrix.shape)
scores_pred_matrix.head(3)
```

```python
from sklearn.metrics import mean_squared_error

# 사용자가 평점을 부여한 영화에 대해서만 예측 성능 평가 MSE 를 구함. 
def get_mse(pred, actual):
#     Ignore nonzero terms.
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten() # 데이터가 많으면 메모리 관리 측면에서 유리
    print(len(pred))
    print(len(actual))
    return mean_squared_error(pred, actual)

print('아이템 기반 모든 인접 이웃 MSE: ', get_mse(scores_pred, scores_matrix.values ))
```

```python
# 사용자별 아이템 유사도에 기반한 예측 평점 데이터 세트 반환하는 함수

def predict_rating_topsim(ratings_arr, item_sim_arr, n=20):
    # 사용자-아이템 평점 행렬 크기만큼 0으로 채운 예측 행렬 초기화
    pred = np.zeros(ratings_arr.shape)
    # 모든 사용자에 대해서 모든 영화의 평점예측을 진행할 초기 행렬

    # 사용자-아이템 평점 행렬의 열 크기만큼 Loop 수행. 
    for col in range(ratings_arr.shape[1]):
        # 코드완성
        
#         if col==2 : # 1번 컬럼에 대해서 수행과정을 확인하고 멈춤(실제함수에서는 제거해야 함)
#             break
#         print(col)
        
        # 유사도 행렬에서 유사도가 큰 순으로 n개 데이터 행렬의 index 반환
        top_n_items = [np.argsort(item_sim_arr[:, col])[:-n-1:-1]]
#         print(top_n_items) # 특정 영화에 대해 유사도가 큰 n개의 영화 index
        
        # 개인화된 예측 평점을 계산(위 수식을 대입)
        for row in range(ratings_arr.shape[0]):
            pred[row, col] = item_sim_arr[col, :][top_n_items].dot(ratings_arr[row, :][top_n_items].T) 
            pred[row, col] /= np.sum(np.abs(item_sim_arr[col, :][top_n_items])) 
    
    return pred
```

```python
import warnings
warnings.filterwarnings("ignore")

scores_pred = predict_rating_topsim(scores_matrix.values , item_sim_df.values, n=20)
print('아이템 기반 인접 TOP-20 이웃 MSE: ', get_mse(scores_pred, scores_matrix.values))


# 계산된 예측 평점 데이터는 DataFrame으로 재생성
scores_pred_matrix = pd.DataFrame(data=scores_pred, index= scores_matrix.index,
                                   columns = scores_matrix.columns)
```

```python
# 선호도를 주지 않은 유기견을 리스트 객체로 반환하는 함수 생성

# ratings_matrix : 실제 사용자 평점 df, 행 : 사용자 id 열 : 영화 평점
def get_unseen_dogs(ratings_matrix, userId):
    # userId로 입력받은 사용자의 모든 유기견 정보 추출하여 Series로 반환
    user_rating = ratings_matrix.loc[userId,:]   
    # 반환된 user_rating 은 유기견(dogId)를 index로 가지는 Series 객체임
    unseen_list = user_rating[user_rating==0].index.to_list()
    
    return unseen_list
```

```python
# 유기견 추천 함수 생성

# 파라미터 : 예측 평점 데이터프레임(ratings_pred_matrix)
# 추천하려는 사용자 id, 추천 후보 유기견 리스트(안 본 유기견), 추천 상위 유기견 수
# 반환 : 사용자가 좋아할만한 가장 높은 예측 선호도을 가진 유기견 추천

def recomm_dog_by_userid(pred_df, userId, unseen_list, top_n=10):
    # 예측 평점 DataFrame에서 사용자id index와 unseen_list로 들어온 영화명 컬럼을 추출하여
    # 가장 예측 평점이 높은 순으로 정렬함. 
    # pred_df.loc[userId, unseen_list] 보지 않은 영화들의 예측평점 data 추출
    recomm_dogs = pred_df.loc[userId, unseen_list].sort_values(ascending=False)[:top_n]
    return recomm_dogs
```
